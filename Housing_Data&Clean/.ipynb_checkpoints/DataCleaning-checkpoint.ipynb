```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import re


# Main dataset read in here


main = pd.read_csv('homevalue_zip.csv')
main = main[main['StateName']=='CA']
main.rename(columns = {"RegionName":"Zip","CountyName":"County"}, inplace=True)

#Check Column Names and their indexes
#i = 0
#for col_name in homevalue.columns: 
    #print(col_name,i)
    #i = i+1

    
# Select the columns we need
    
main_subset1 = main.iloc[:,177:312]
main_subset2 = main.iloc[:,[8,2,5,6]]


main = pd.concat([main_subset1, main_subset2], axis=1)
main
```

```python
# Density and crime rate data read in here
crime = pd.read_csv('crime&population_density.csv')
crime.rename(columns = {"Populationdensity[5][3][note 2]":"Population_density","Property crime rateper 1,000 persons":"Crime_rate_Per1000","City/Agency":"City"}, inplace=True)
crime = crime[["Population_density","Crime_rate_Per1000","County","City"]]

# First merge

main_crime = pd.merge(main,crime,how = "left",on = "City")
main_crime
```

```python
# Read in high school ranking data
# Note here I had to get the mean grouping by city and by county because I noticed that some of the city does
# not have a school so for those I used the average school rating from the county the city belongs to
# This is the best way I can think of dealing with the NA values for school ranking

school = pd.read_excel('California_highschool_ranking.xlsx')
school = school.drop(labels=[0,1,2],axis=0)
school = school.rename(columns=school.iloc[0])
school = school.drop(labels=[3],axis=0)
school = school[["SchoolDigger Star Rating","School","County","Zip","City"]]
school["SchoolDigger Star Rating"] = school["SchoolDigger Star Rating"].astype(int)



School_rating_ave_byCity = school.groupby(["City"])["SchoolDigger Star Rating"].mean()
School_rating_ave_byCity = School_rating_ave_byCity.reset_index()


School_rating_ave_byCounty = school.groupby(["County"])["SchoolDigger Star Rating"].mean()
School_rating_ave_byCounty = School_rating_ave_byCounty.reset_index()
```

```python
# Merge school ranking data with main df
# Fillna method was pretty handy to use

main_crime_school= pd.merge(main_crime,School_rating_ave_byCity,how = "left",on = "City")
main_crime_school = main_crime_school.drop(columns = "County_y").rename(columns = {"County_x":"County"})


main_crime_school= pd.merge(main_crime_school,School_rating_ave_byCounty,how = "left",on = "County")
main_crime_school["SchoolDigger Star Rating_x"] = main_crime_school["SchoolDigger Star Rating_x"].fillna(main_crime_school['SchoolDigger Star Rating_y'])


main_crime_school = main_crime_school.drop(columns = "SchoolDigger Star Rating_y").rename(columns = {"SchoolDigger Star Rating_x":"School_ranking"})

# Here I started to try to figure out how to label the crime rate and the population density by quantiles, there were couple things
# I could not figure out. I tried for about any hour, so I decided to use R to deal with this, I also pushed the Rmd file onto Github


main_crime_school.to_csv("main_crime_school.csv")
```
